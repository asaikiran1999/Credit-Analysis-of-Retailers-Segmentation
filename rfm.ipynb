```python
import pandas as pd
df = pd.read_csv("/content/CreditAnalysis_data.xlsx - CreditAnalysis_data.csv")
```

```python
df.head(10)
```

```python
df['ordereditem_product_id'] = df['ordereditem_product_id'].fillna(0)
df.head(10)
```

```python
list1 = df['retailer_names'].unique()
```

```python
df.shape[0]
```

```python
list2=[]
for i in list1:
  count = 0
  for j in range(df.shape[0]):
    if df['retailer_names'][j]==i:
      count+=1
  list2.append(count)
```

```python
merged_list_count = tuple(zip(list1, list2))
merged_list_count
```

```python
df['frequency']=0
for i in merged_list_count:
  for j in range(df.shape[0]):
    if df['retailer_names'][j]==i[0]:
      df['frequency'][j]=i[1]
```

```python
df.head()
```

```python
df['recency']= df['created'].rank()
```

```python
df.head()
```

```python
list2=[]
for i in list1:
  count = 0
  for j in range(df.shape[0]):
    if df['retailer_names'][j]==i:
      count+=df['value'][j]
  list2.append(count)
```

```python
merged_list_value = tuple(zip(list1, list2))
```

```python
df['monetary']=0
for i in merged_list_value:
  for j in range(df.shape[0]):
    if df['retailer_names'][j]==i[0]:
      df['monetary'][j]=i[1]
```

```python
df.head()
```

```python
x=list(zip(*merged_list_value))
x=list(x[1])
x.sort(reverse = True)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Read CSV into pandas
list1=[]
list2=[]
for i in merged_list_value:
  for j in x:
    if(i[1]==j):
      list1.append(i[0])

name = list1[0:10]
price = x[0:10]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(name, price, color ='blue',
        width = 0.2)
plt.xticks(rotation='vertical')
plt.xlabel("x")
plt.ylabel("y")
plt.title("top retailers according to monetary")
plt.show()
```
